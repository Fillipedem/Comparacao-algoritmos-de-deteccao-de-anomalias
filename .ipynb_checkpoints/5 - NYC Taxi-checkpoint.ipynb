## Preprocessamento

```python
df_raw, labels = loader.load('nyc_taxi.csv')
df_raw.head()
```

Valores **duplicados** e **missing data**

```python
df_raw.index.duplicated().sum()
```

```python
drange = pd.date_range(start=df_raw.index.min(), end=df_raw.index.max(), freq='30T')
drange
```

**Visualização**

```python
plot_anomalies(df_raw['value'], windows=labels)

plt.title("Janela de anomalia",fontsize=20);
```

## Algoritmos

### MAD

```python
ts = df_raw['value']
```

```python
mad = MAD()
mad.fit(ts)
madpredict = mad.predict(ts)
madpredict.head()
```

```python
# score por dia
mad_dayscore = madpredict.resample('D').count() .dropna()/24
```

```python
tmpdf = pd.DataFrame(np.zeros(len(drange)), index=drange)
mad_dayscore = (tmpdf[0] + mad_dayscore).fillna(0)
mad_dayscore
```

```python
plot_anomalies(mad_dayscore, windows=labels)

plt.title("MAD score - Dia count",fontsize=20)
plt.ylabel('Score de anomalia');
```

### Tukey

```python
ts = df_raw['value']
```

```python
tu = Tukey()
tu.fit(ts)
tupredict = tu.predict(ts)
tupredict.head()
```

```python
tu_dayscore = tupredict.resample('D').count().dropna()/24

tmpdf = pd.DataFrame(np.zeros(len(drange)), index=drange)
tu_dayscore = (tmpdf[0] + tu_dayscore).fillna(0)
```

```python
plot_anomalies(tu_dayscore, windows=labels)
plt.title("Tukey score - Count do Dia",fontsize=20)
plt.ylabel('Score de anomalia', fontsize=15);
```

### Twitter

```python
ts = df_raw['value']
```

```python
twitterpredict = twitter(ts)
twitterpredict.head()
```

```python
twitter_dayscore = twitterpredict.resample('D').count().dropna()/24
```

```python
plot_anomalies(twitter_dayscore, windows=labels)
plt.title("Twitter score",fontsize=20)
plt.ylabel('Score de anomalia');
```

```python
from statsmodels.tsa.seasonal import STL

# plot size
plt.rcParams['figure.figsize'] = (20, 10)
# seasonal decompose
stl = STL(ts, period=12)
results = stl.fit()
results.plot();
```
