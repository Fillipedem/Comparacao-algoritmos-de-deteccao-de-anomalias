## Algoritmos

### MAD

```python
ts = df['value']
```

```python
mad = MAD()
mad.fit(ts)
madpredict = mad.predict(ts)
madpredict.head()
```

```python
# score por dia
mad_dayscore = madpredict.resample('D').count().dropna()/24
```

```python
plot_anomalies(mad_dayscore, windows=labels)

plt.title("MAD score - Dia count",fontsize=20)
plt.ylabel('Score de anomalia');
```

### Tukey

```python
ts = df['value']
```

```python
tu = Tukey()
tu.fit(ts)
tupredict = tu.predict(ts)
tupredict.head()
```

```python
tu_dayscore = tupredict.resample('D').count().dropna()/24
```

```python
plot_anomalies(tu_dayscore, windows=labels)
plt.title("Tukey score - Count do Dia",fontsize=20)
plt.ylabel('Score de anomalia', fontsize=15);
```

**PACF**

```python
plot_pacf(ts_diff, lags=40);
```

```python
def print_afduller(ts):
    result = adfuller(ts)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
                              
print_afduller(ts_diff)
```

##### Training

```python
def arima_forecast(train, test, order=(1, 1, 1)):
    ''' forescast one step ahead on the test'''
    x = ts.values
    history = [h for h in train]
    predictions = []
    conf_inter = []
    for t in range(len(test)):
        print('{}/{}'.format(t, len(test)))
        model = ARIMA(history, order=order)
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        predictions.append(output[0])# predict
        conf_inter.append(output[2]) # conf_inter
        history.append(test[t])
    return predictions, conf_inter
```

```python
def split(x, size=0.66):
    split_point = int(len(x)*size)
    return x[:split_point], x[split_point:]

train, test = split(ts)
train.shape, test.shape
```
